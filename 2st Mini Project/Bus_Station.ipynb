#### ARS 번호가 object인 이유를 찾아내고 해결

* ARS번호가 object인 이유는 '~' 이 들어갔기 때문이다.
* 버스정류장ARS번호 중 '~'로 표시되는 정류장은 '가상'(기점가상/종점가상/가상) 정류장이다

* 가상버스정류장은 가상으로 보이지 않는 버스정류장을 만들어서 이를 기반으로 버스의 실시간 위치정보 제공에 활용할 수 있습니다.

```python
# ARS 번호가 object인 이유를 찾아내고 해결 - '가상 버스정류장'

bus[bus['역명'].str.contains('가상')]
```

```python
bus[bus['버스정류장ARS번호'] == '~']
```

* 데이터 안에 있는 경기도 지역의 버스 정류장 걸러내기
* 외부데이터 : 서울시 버스정류소 위치정보

```python
location = pd.read_csv('./서울시 버스정류소 위치정보.csv', encoding = 'cp949')
location
```

```python
location.rename(columns={'정류소번호' : '버스정류장ARS번호'}, inplace = True)
```

```python
# merge 정류장 번호 > 서울시만 해당
df = pd.merge(bus, location , how = 'inner', on = '버스정류장ARS번호')
```

```python
df = df.drop(['X좌표', 'Y좌표', '정류소명'], axis = 1)
```

```python
# 서울시 전체에 있는 버스 정류장 총 개수
df['버스정류장ARS번호'].nunique()
```

* 노선번호는 같지만 노선명이 다른 경우가 있을 수 있다.

```python
# '구' 별로 정류장 개수를 파악
bus_station_station = df.groupby(by = ['자치구'], as_index = False)['버스정류장ARS번호'].nunique()
bus_station_station.columns = ['자치구', '정류장 개수']
```

```python
# '구' 별로 버스 노선개수를 파악
bus_station_line = df.groupby(by = ['자치구'], as_index = False)['노선번호'].nunique()
bus_station_line.columns = ['자치구', '정류장 노선개수']
```

```python
# '구' 별로 승차 총 승객수, 하차 총 승객수 확인하기
bus_station_inout_sum = df.groupby(by = ['자치구'], as_index = False)[['승차총승객수','하차총승객수']].sum()
```

```python
# '구' 별로 승차 평균 승객수, 하차 평균 승객수 확인하기
bus_station_inout_mean = df.groupby(by = ['자치구'], as_index = False)[['승차총승객수','하차총승객수']].mean()
bus_station_inout_mean.columns = ['자치구', '정류장 승차 평균 승객수', '정류장 하차 평균 승객수']
```

```python
bus_sol = pd.merge(bus_station_station, bus_station_line, on = '자치구', how = 'inner')
bus_sol = pd.merge(bus_sol, bus_station_inout_sum, on = '자치구', how = 'inner')
bus_sol = pd.merge(bus_sol, bus_station_inout_mean, on = '자치구', how = 'inner')
bus_sol
```
