#### 실습문제 7. '도착시군구코드'를 기준으로 각 구별로 평균 이동시간(분)의 평균과 합을 구하기

* groupby 사용, as_index=False로 설정해주세요

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_moving_sum = df.groupby(by=['도착 시군구 코드'], as_index= False)[['평균 이동 시간(분)','이동인구(합)']].sum().sort_values(by='도착 시군구 코드', ascending =False)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_moving_mean = df.groupby(by=['도착 시군구 코드'], as_index= False)[['평균 이동 시간(분)','이동인구(합)']].mean().sort_values(by='도착 시군구 코드', ascending =False)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_moving_sum.rename(columns={"평균 이동 시간(분)":'총 이동 시간', '이동인구(합)':'총 이동 인구'},inplace=True)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_moving_mean.rename(columns={"평균 이동 시간(분)":'평균 이동 시간', '이동인구(합)':'평균 이동 인구'},inplace=True)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_seoul_moving = pd.merge(seoul_moving_sum,seoul_moving_mean, how = 'inner', on = '도착 시군구 코드')
df_seoul_moving
```

#### 실습문제 9. 어느 요일의 심야 이동이 가장 많을까요?
* 심야 기준 : 00시~05시

```python
# 0:00~ 4:59 데이터 불러오기 

seoul_moving_0 = pd.read_csv('생활이동_자치구_2022.12_00시.csv', sep=",", encoding = 'cp949')
seoul_moving_1 = pd.read_csv('생활이동_자치구_2022.12_01시.csv', sep=",", encoding = 'cp949')
seoul_moving_2 = pd.read_csv('생활이동_자치구_2022.12_02시.csv', sep=",", encoding = 'cp949')
seoul_moving_3 = pd.read_csv('생활이동_자치구_2022.12_03시.csv', sep=",", encoding = 'cp949')
seoul_moving_4 = pd.read_csv('생활이동_자치구_2022.12_04시.csv', sep=",", encoding = 'cp949')
```

```python
# 0:00~ 4:59 데이터 합치기 
seoul_night = pd.concat([seoul_moving_0,seoul_moving_1,seoul_moving_3,seoul_moving_4], axis = 0)
seoul_night
```

```python
# 결측치 처리하기
seoul_night.replace("*","1.5", inplace=True)
```

```python
# 데이터 타입 변경
seoul_night = seoul_night.astype({'이동인구(합)':'float'})
seoul_night.info()
```

```python
# 서울 지역의 데이터만 골라보기
seoul_night = seoul_night[(seoul_night["출발 시군구 코드"] < 20000) & (seoul_night["도착 시군구 코드"] < 20000)]
seoul_night
```

```python
# 심야 시간대 어느 요일의 이동이 가장 많을까?
seoul_moving_mean = seoul_night.groupby(by=['요일'], as_index= False)['이동인구(합)'].mean().sort_values(by='이동인구(합)', ascending =False)
```
