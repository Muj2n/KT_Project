#### 실습문제 4. 가장 패스가 많았던 나라를 구하시오 그리고 나라별 패스 정확도를 구하시오 

```python
passes = df[['team1','team2', 'passes team1', 'passes team2', 'passes completed team1', 'passes completed team2']]
```

```python
passes
```

```python
pass1 = passes.groupby('team1', as_index = False).agg({'passes team1':'sum','passes completed team1':'sum'})
# agg : 매핑함수 - 여러개의 열에 여러가지 함수를 적용할 수 있다.
# group객체.agg([함수1,함수2,함수3,…]) / group객체.agg({‘열1’: 함수1, ‘열2’:함수2, …})
# 예) .agg({'fare':['min','max'], 'age':'mean'})
pass2 = passes.groupby('team2', as_index = False).agg({'passes team2':'sum','passes completed team2':'sum'})
```

```python
 passes.groupby('team1', as_index = False).agg({'passes team1':['sum','count'],'passes completed team1':'sum'})
```

```python
# 정확도 (%)
pass1['pass accuracy team1'] = pass1['passes completed team1']/pass1['passes team1'] * 100
pass2['pass accuracy team2'] = pass2['passes completed team2']/pass2['passes team2'] * 100
```

```python
total_passes = pass1.merge(pass2, left_on = 'team1', right_on='team2',how='left')
```

```python
total_passes['total_passes'] = total_passes['passes team1'] + total_passes['passes team2']
total_passes['total_completed_passes'] = total_passes['passes completed team1'] + total_passes['passes completed team2']
```

```python
total_passes['total_accuracy'] = total_passes['total_completed_passes'] / total_passes['total_passes'] * 100
```

```python
total_passes.head()
```

```python
total_passes.sort_values(by='total_passes',ascending=False).head()
```

#### 실습문제 5. 나라별 평균 점유율과 경기당 평균 패스 수를 구해보시오

```python
possession = df[['team1','team2', 'possession team1',	'possession team2', 'passes team1',	'passes team2']]
```

```python
possession
```

```python
# % 제거
possession['possession team1'] = df['possession team1'].str.replace("%",'')
possession['possession team2'] = df['possession team2'].str.replace("%",'')
```

```python
# 데이터 숫자 형식으로 변환 = astype(int)
possession['possession team1'] = pd.to_numeric(possession['possession team1'])
possession['possession team2'] = pd.to_numeric(possession['possession team2'])
```

```python
possession1 = possession.groupby('team1', as_index = False).agg({'passes team1':'mean','possession team1':'mean'})
possession2 = possession.groupby('team2', as_index = False).agg({'passes team2':'mean','possession team2':'mean'})
```

```python
total_possession = possession1.merge(possession2, left_on = 'team1', right_on='team2',how='left')
```

```python
total_possession['pass_mean'] = (total_possession['passes team1'] + total_possession['passes team2'])/2
total_possession['possession_mean'] = (total_possession['possession team1'] + total_possession['possession team2'])/2
```

```python
total_possession.head()
```

#### 실습문제 6. 나라별로 파울의 수와 받은 레드카드와 옐로카드 수를 구하시오

```python
foul = df[['team1','team2', 'yellow cards team1', 'yellow cards team2',	'red cards team1', 'red cards team2', 'fouls against team1', 'fouls against team2']]
```

```python
foul1 = foul.groupby('team1',as_index=False).agg({'yellow cards team1':'sum','red cards team1':'sum',"fouls against team1":'sum'})
foul2 = foul.groupby('team2',as_index=False).agg({'yellow cards team2':'sum','red cards team2':'sum',"fouls against team2":'sum'})
```

```python
fouls = foul1.merge(foul2,left_on="team1",right_on="team2")
```

```python
fouls['yellow cards sum'] = fouls['yellow cards team1'] + fouls['yellow cards team2']
fouls['red cards sum'] = fouls['red cards team1'] + fouls['red cards team2']
fouls['fouls against team sum'] = fouls['fouls against team1'] + fouls['fouls against team2']
```

```python
fouls.sort_values(by='fouls against team sum',ascending=False).head()
```

#### 실습문제 7. 나라별로 슈팅 시도 횟수를 구하시오

```python
attempts_on_1=df.groupby('team1',as_index=False)['on target attempts team1'].sum()
attempts_on_2=df.groupby('team2',as_index=False)['on target attempts team2'].sum()
```

```python
attempts = attempts_on_1.merge(attempts_on_2,left_on="team1",right_on="team2")
```

```python
attempts['total_attempts'] = attempts['on target attempts team1'] + attempts['on target attempts team2']
```

```python
attempts.sort_values(by='total_attempts',ascending=False)
```
